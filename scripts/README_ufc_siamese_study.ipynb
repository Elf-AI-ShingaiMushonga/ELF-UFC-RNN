# UFC Siamese Study Pipeline

This runner operationalizes the paper draft into executable experiments:

- data loading and leakage-safe temporal split
- baselines: naive pre-fight win-rate, logistic regression, gradient boosting, MLP
- Siamese RNN (shared GRU fighter encoder)
- calibration tables
- subgroup analysis
- ablation study variants
- artifact export for paper tables/figures


## Install ML dependencies

```bash
pip install -r requirements-ml.txt
```


## Run full study

```bash
python3 scripts/run_ufc_siamese_study.py \
  --input-csv data/ufc_fights_rnn.csv \
  --output-dir outputs \
  --run-name paper_run_01
```


## Key runtime options

```bash
python3 scripts/run_ufc_siamese_study.py \
  --max-seq-len 10 \
  --epochs 35 \
  --batch-size 128 \
  --hidden-dim 128 \
  --device auto \
  --num-workers 4 \
  --lr 0.0007 \
  --seed 42
```

Skip ablation runs (faster):

```bash
python3 scripts/run_ufc_siamese_study.py --skip-ablations
```


## Outputs

Each run writes to `outputs/<run-name>/`:

- `main_metrics.csv`: core model comparison table
- `test_predictions.csv`: per-fight predictions on test split
- `calibration_<model>.csv`: reliability-bin tables
- `subgroup_<model>.csv`: subgroup performance slices
- `ablation_metrics.csv`: ablation study results (if enabled)
- `siamese_training_history.csv`: epoch history
- `run_metadata.json`: split boundaries, feature counts, config


## Notes

- The script requires binary labels (`fighter_1_win` / `fighter_2_win`) and drops non-binary outcomes.
- Temporal split is event-date ordered to reduce look-ahead bias.
- Sequence inputs are built chronologically so each sample only uses prior fights.
